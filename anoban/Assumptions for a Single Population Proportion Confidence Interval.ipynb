```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# ___Assumptions for a Single Population Proportion Confidence Interval___
----------------

```python
# consifidence interval = best estimate +- margin of error
# margin of error = multiplier * standard error
```

```python
# When calculating the confidence intervals, we need to make sure the sample meets certain criteria:
# 1) The sample must be a simple random sample -> representative of the population
# 2) Sample is large enough.
```

```python
# If the sample is large enough, we can approximate the sampling distribution to a normal distribution.
# The multiplier in the margin of error computation is taken from a standard normal distribution.
```

```python
# How do we define "large enough"?

# As the sample size increases, the sampling curve will become more and more closer to a bell shaped one.
```

```python
data = pd.read_csv("../data/EV_Population_Data.csv")
```

```python
data["isHonda"] = data.Make == "HONDA"
```

```python
data.isHonda.mean()
```

```python
data.shape
```

```python
limit = data.shape[0]
ssizes = [100, 200, 500, 1000, 2000, 5000, 7500, 10000, 25000]
fig, axes = plt.subplots(nrows = 3, ncols = 3, sharex = False, sharey = False)
fig.set_size_inches(20, 20)
axes = axes.ravel()

for (x, ssize) in enumerate(ssizes):
    prop_honda = []
    for i in range(1000):
        prop_honda.append(data.isHonda[np.random.randint(0, limit, size = ssize)].mean())
    axes[x].hist(prop_honda, bins = 200, rwidth = 0.9)
    axes[x].set_title(f"Sample size {ssize:,}")

plt.show()
```
