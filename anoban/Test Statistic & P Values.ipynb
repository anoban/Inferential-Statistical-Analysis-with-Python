```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# ___Test Statistic & P Values___
---------------------

```python
# The test statistic (in the case of proportions, it'll be a z score)
# follows a standard normal distribution, i.e a normal distribution with mean = 0 and std = 1

# So a standard normal distribution will form a bell shaped curve;

stdnorm = np.random.standard_normal(size = 1_000_000)

fig, axes = plt.subplots()
axes.hist(stdnorm, bins = 200, color = "red", alpha = 0.75)
plt.show()
```

```python
# Suppose we get a z score of 0.879

z = 0.879

fig, axes = plt.subplots()
axes.hist(stdnorm, bins = 200, color = "red", alpha = 0.75)
axes.vlines(x = z, ymin = 0, ymax = 20_000, color = "black")
plt.show()
```

```python
# Now we have the Z score in our distribution,
# Then to compute the p value, we need to decide which side of the line to shade (or calculate the area for)
# This is determined by the alternative hypothesis.
```

```python
# Let's say that our null hypothesis equalted a metric to 87.6672
# Alternative hypothesis claimed that metric is less than 87.6672
```
